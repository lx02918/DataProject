```python
# 2 两数相加
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        curr = dummy
        carry = 0
        while l1 or l2 or carry:
            val1 = val2 = 0
            if l1:
                val1 = l1.val
                l1 = l1.next
            if l2:
                val2 = l2.val
                l2 = l2.next
            carry, val = divmod(val1 + val2 + carry, 10)
            curr.next = ListNode(val)
            curr = curr.next
        return dummy.next
```

```python
# 21 合并两个有序链表
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        curr = dummy
        while list1 and list2:
            if list1.val < list2.val:
                curr.next = list1
                list1 = list1.next
            else:
                curr.next = list2
                list2 = list2.next
            curr = curr.next
        curr.next = list1 or list2
        return dummy.next
```

```python
# 138 随机链表的复制
"""
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
"""

class Solution:
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        if not head:
            return None
        curr = head
        dic = {}
        while curr:
            dic[curr] = Node(curr.val)
            curr = curr.next
            
        curr = head
        while curr:
            dic[curr].next = dic.get(curr.next)
            dic[curr].random = dic.get(curr.random)
            curr = curr.next
        return dic.get(head)
```

```python
# 92 反转链表 II
class Solution:
    def reverseBetween(self, head: Optional[ListNode], m: int, n: int) -> Optional[ListNode]:
        dummy = ListNode(0)
        dummy.next = head
        pre = dummy
        for _ in range(m - 1):
            pre = pre.next
        start = pre.next
        then = start.next
        for _ in range(n - m):
            start.next = then.next
            then.next = pre.next
            pre.next = then
            then = start.next
        return dummy.next 
```

```python
# 25 K 个一组翻转链表
class Solution:
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if not head:
            return None
        a = b = head
        for _ in range(k):
            if not b:
                return head
            b = b.next  
        newHead = self.reverse(a, b)
        a.next = self.reverseKGroup(b, k)
        return newHead
    
    def reverse(self, a, b):
        prev = None
        curr = a
        while curr != b:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        return prev
```

```python
# 19 删除链表的倒数第 N 个节点
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dummy = ListNode(0)
        dummy.next = head
        cur, length = head, 0
        # 获取长度
        while cur:
            length += 1
            cur = cur.next
        cur = dummy
        # 找到倒数第 n 个节点
        for _ in range(length - n):
            cur = cur.next
        # 删除节点
        cur.next = cur.next.next
        return dummy.next
```

```python
# 82 删除排序链表中的重复元素 II
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head:
            return None
        dummy = ListNode(0)
        dummy.next = head
        curr = dummy
        while curr.next and curr.next.next:
            if curr.next.val == curr.next.next.val:
                duplicate_val = curr.next.val
                while curr.next and curr.next.val == duplicate_val:
                    curr.next = curr.next.next
            else:
                curr = curr.next
        return dummy.next
```

```python
# 61 旋转链表
class Solution:
    def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if not head:
            return None
        curr = head
        count = 0
        while curr:
            count += 1
            curr = curr.next
        k = k % count
        if k == 0:
            return head
        fast = slow = head
        for _ in range(k):
            fast = fast.next
        while fast.next:
            slow = slow.next
            fast = fast.next
        new_head = slow.next
        slow.next = None
        fast.next = head
        return new_head
```

```python
# 86 分隔链表
class Solution:
    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        smaller = ListNode(0)
        bigger = ListNode(0)
        small, big = smaller, bigger
        while head:
            if head.val < x:
                small.next = head
                small = small.next
            else:
                big.next = head
                big = big.next  
            head = head.next
        small.next = bigger.next
        big.next = None
        return smaller.next
```

```python
# 146 LRU缓存机制
class CreatNode:

    def __init__(self, key = 0, value = 0):
        self.key = key
        self.value = value
        self.prev = None
        self.next = None

class LRUCache:

    def __init__(self, capacity: int):
        self.hash = {}
        self.head = CreatNode()
        self.tail = CreatNode()
        self.head.next = self.tail
        self.tail.prev = self.head
        self.capacity = capacity
    def get(self, key: int):
        if key not in self.hash:
            return -1
        node = self.hash[key]
        self._remove(node)
        self._add(node)
        return node.value
    

    def put(self, key: int, value: int) -> None:
        if key in self.hash:
            node = self.hash[key]
            node.value = value
            self._remove(node)
            self._add(node)
        else:
            node = CreatNode(key, value)
            self._add(node)
            self.hash[key] = node
            if len(self.hash) > self.capacity:
                del self.hash[self.head.next.key]
                self._remove(self.head.next)

    def _remove(self, node):
        prev = node.prev
        next = node.next
        prev.next = next
        next.prev = prev

    def _add(self, node):
        prev = self.tail.prev
        prev.next = node
        node.prev = prev
        node.next = self.tail
        self.tail.prev = node
```
