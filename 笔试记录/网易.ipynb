```python
# 第一题
class Solution:
    def countWords(self , s: str) -> int:
        words = s.replace(',', ' ').replace('!', ' ').split()
        unique_words = set(word.lower() for word in words if not word.isdigit())
        return len(unique_words)
```

```python
# 第二题
from functools import cmp_to_key

def smallest_number(nums):
    if not nums:
        return ''
    negative = next((num for num in nums if num < 0), None)
    positive = [str(num) for num in nums if num >= 0]
    
    def compare(a, b):
        if a + b > b + a:
            return 1
        elif a + b < b + a:
            return -1
        else:
            return 0
    positive.sort(key = cmp_to_key(compare))
    
    result = ''.join(positive)
    if negative:
        result = str(negative) + result
    return result if result else '0'
nums = list(map(int, input().split()))
print(smallest_number(nums))
```

```python
# 第二题
from functools import cmp_to_key
def minInteger(self, nums: List[int]) -> str:
    def sort_rule(x, y):
        a, b = x + y, y + x
        if a < b: return 1
        elif a > b: return -1
        else: return 0
    strs = [str(num) for num in nums]
    strs.sort(key = cmp_to_key(sort_rule))
    if strs[0] == '0':
        return '0'
    return ''.join(strs)
nums = list(map(int, input().split()))
print(minInteger(nums))
```

```python
# 第三题
class Solution:
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        if len(edges)!= n - 1:
            return False
        parent = list(range(n))
        def find(x):
            if parent[x] != x:
                parent[x] = find(parent[x])
            return parent[x]
        def union(x, y):
            parent[find(x)] = find(y)
        for a, b in edges:
            if find(a) == find(b):
                return False
            union(a, b)
        return len(set(find(x) for x in range(n))) == 1
```

```python
import sys
import math
def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0] **2 + (p1[1] - p2[1]) **2))
def brute_force(points):
    n = len(points)
    if n < 2:
        return None, float('inf')
    min_dist = float('inf')
    pair = None
    for i in range(n):
        for j in range(i + 1, n):
            dist = distance(points[i], points[j])
            if dist < min_dist:
                min_dist = dist
                pair = (points[i], points[j])
    return pair, min_dist
def strip_closest(strip, d):
    strip.sort(key = lambda point: point[1])
    min_dist = d
    pair = None
    for i in range(len(strip)):
        j = i + 1
        while j < len(strip) and (strip[j][i] - strip[i][i]) < min_dist:
            dist = distance(strip[i], strip[j])
            if dist < min_dist:
                min_dist = dist
                pair = (strip[i], strip[j])
            j += 1
    return pair, min_dist
def closest_pair(points):
    n = len(points)
    if n <= 3:
        return brute_force(points)
    
    mid = n // 2
    mid_points = points[mid]
    left = points[:mid]
    right = points[mid:]
    
    left_pair, left_dist = closest_pair(left)
    right_pair, right_dist = closest_pair(right)
    
    if left_pair < right_pair:
        d = left_dist
        pair = left_pair
    else:
        d = right_dist
        pair = right_pair
    strip = [p for p in points if abs(p[0] - mid_points[0]) < d]
    strip_pair, strip_dist = strip_closest(strip, d)
    if strip_dist < d:
        return strip_pair, strip_dist
    else:
        return pair, d
def main():
    input_str = input().strip()
    points = eval(input_str)
    points.sort()
    closest_points, _ = closest_pair(points)
    if closest_points[0] <= closest_points[1]:
        print(f"(({closest_points[0][0]}, {closest_points[0][1]}), ({closest_points[1][0]}, {closest_points[1][1]}))")
    else:
        print(f"(({closest_points[1][0]}, {closest_points[1][1]}), ({closest_points[0][0]}, {closest_points[0][1]}))")
    if __name__ == '__main__':
        main()
```

```python
# 第四题
import sys
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
def distance(a, b):
    return math.sqrt(a.x - b.x) ** 2 + (a.y - b.y) ** 2
def min_length(points, left, right):
    if left == right:
        return float('inf')
    if left + 1 == right:
        return distance(points[left], points[right])
    mid = (left + right) // 2
    d1 = min_length(points, left, mid)
    d2 = min_length(points, mid, right)
    min_dist = min(d1, d2)
    
    strip = [p for p in points[left:right + 1] if abs(p.x - points[mid].x) <= min_dist]
    strip.sort(key = lambda p: p.y)
    
    for i in range(len(strip)):
        j = i + 1
        while j < len(strip) and (strip[j].y - strip[i].y) < min_dist:
            min_dist = min(min_dist, distance(strip[i], strip[j]))
            j += 1
    return min_dist
def main():
    n = int(input())
    points = []
    for _ in range(n):
        x, y = map(int, input().split())
        points.append(Point(x, y))
        points.sort(key = lambda p: (p.x, p.y))
        result = min_length(points, 0, n - 1)
        print(f"{result:.3f}")
if __name__ == '__main__':
    main()
```

```python
import copy
def distance(a, b):
    return (b[0] - a[0]) ** 2 + (b[1] - a[1]) ** 2
def find_nearest_points(points_x, points_y, n):
    if n == 2:
        return [distance(points_x[0], points_y[1]), (points_x[1], points_y[0])]
    if n == 1:
        return [999, (points_x[0], points_x[0])]
    left_points = points_x[0:int(n/2)]
    right_points = points_x[int(n/2):n + 1]
    lx = (points_x[int(n/2)][0] + points_x[int(n/2) - 1][0]) /2
    d1 = find_nearest_points(left_points, points_y, int(n/2))
    d2 = find_nearest_points(right_points, points_y, int(n/2) + (n % 2))
    min_set = min(d1, d2, key = lambda x: x[0])
    min_points = min_set[1]
    
    T = []
    for i in points_y:
        if lx - min_set[0] <= i[0] <= lx + min_set[0]:
            T.append(i)
    min_dis = min_set[0]
    for k in T:
        for i in range(1, 5):
            if T.index(k) + i < len(T) - 1:
                if T[T.index(k) + i][1] - k[1] < min_set[0]:
                    dis = distance(k, T[T.index(k) + i])
                    if dis < min_dis:
                        min_dis = dis
                        min_points[0] = k
                        min_points[1] = T[T.index(k) + i]
                        
                else:
                    break
                    
            else:
                break
    return [min(min_dis, min_set[0]), min_points]
```
