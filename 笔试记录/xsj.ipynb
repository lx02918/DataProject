```python
class Solution:
    def largestRectangleArea(self , heights: List[int]) -> int:
        stack = []
        max_area = 0
        heights.append(0)
        for i in range(len(heights)):
            while stack and heights[stack[-1]]:
                h = heights[stack.pop()]
                w = i if not stack else i - stack[-1] - 1
                max_area = max(max_area, h * w)
            stack.append(i)
        return max_area
```

```python
class Solution:
    def largestRectangleArea(self , heights: List[int]) -> int:
        stack = []
        heights = [0] + heights + [0]
        max_area = 0
        for i in range(len(heights)):
            while stack and heights[stack[-1]] > heights[i]:
                h = heights[stack.pop()]
                w = i if not stack else i - stack[-1] - 1
                max_area = max(max_area, h * w)
            stack.append(i)
        return max_area
```

```python
class Solution:
    def compressString(self , s: str, k: int) -> int:
        n = len(s)
        dp = [[float('inf') * (k + 1) for _ in range(n + 1)]]
        dp[0][0] = 0
        for i in range(1, n + 1):
            for j in range(0, k + 1):
                if dp[i - 1][j] < float('inf'):
                    if j + 1 <= k:
                        dp[i][j + 1] = min(dp[i][j + 1], dp[i - 1][j])
                    cnt = 0
                    for m in range(i, n + 1):
                        if s[m - 1] == s[i - 1]:
                            cnt += 1
                            cost = 1
                            if cnt >= 100:
                                cost += 3
                            elif cnt >= 10:
                                cost += 2
                            elif cnt >= 2:
                                cost += 1
                            if j + (m - i) <= k:
                                dp[m][j + (m - i)] = min(dp[m][j + (m - i)], dp[i - 1][j] + cost)
                        else:
                            continue
        return min(dp[n])
```

```python
class Solution:
    def compressString(self , s: str, k: int) -> int:
        n = len(s)
        memo = {}
        def compress(idx, deleted):
            if idx == n:
                return 0
            if (idx, deleted) in memo:
                return memo[(idx, deleted)]
            
            deleted_option = float('inf')
            if deleted < k:
                deleted_option = compress(idx + 1, deleted + 1)
            keep_option = 1
            count = 1
            for j in range(idx + 1, n):
                if s[j] == s[idx]:
                    count += 1
                else:
                    break
            if count > 1:
                keep_option += len(str(count))
            keep_option += compress(idx + count, deleted)
            result = min(deleted_option, keep_option)
            memo[(idx, deleted)] = result
            return result
        
        return compress(0, 0)
```

```python
class Solution:
    def compressString(self , s: str, k: int) -> int:
        def compress(s):
            if not s:
                return 0
            compressed = []
            count =1
            for i in range(1, len(s)):
                if s[i] == s[i-1]:
                    count += 1
                else:
                    compressed.append(s[i-1])
                    if count > 1:
                        compressed.append(str(count))
                    count = 1
            compressed.append(s[-1])
            if count > 1:
                compressed.append(str(count))
            return len(''.join(compressed))
        def dfs(index, k, current):
            if index == len(s):
                return compress(current)
            keep = dfs(index + 1, k, current + s[index])
            deleted = float('inf')
            if k > 0:
                deleted = dfs(index + 1, k - 1, current)
            return min(keep, deleted)
        return dfs(0, k, '')
```
